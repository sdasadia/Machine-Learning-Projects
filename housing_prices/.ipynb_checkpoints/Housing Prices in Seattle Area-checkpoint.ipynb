## In this case study, the housing prices in the Seattle, WA are studied. Graphlab package will be used. The goal is to bild a model to predict house price.

## Fire up graphlab create and loas house sales data

```python
import graphlab
```

```python
data = graphlab.SFrame('home_data.gl/')
```

```python
data
```

## Exploring the data for housing sales

### First consider the case where house price is correlated with the area of the living space.

```python
graphlab.canvas.set_target('ipynb') # Set the ipnb as the target for output
```

```python
data.show(view = "Scatter Plot", x = "sqft_living", y = "price")
```

```python
training,testing = data.random_split(0.6,seed = 15) # Create test and training data set
```

## Build a simple linear model that correlates housing price and living space

```python
sqft_model = graphlab.linear_regression.create(training, target = "price", features = ["sqft_living"],validation_set=None)
```

```python
print testing['price'].mean()
```

```python
print sqft_model.evaluate(testing)
```

## Plot data with prediction model using matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# to see output in the notebook
```

```python
plt.plot(testing['sqft_living'],testing['price'],'*',
        testing['sqft_living'],sqft_model.predict(testing),'-')
```

### The blue * are original data and green line is the prediction from the simple living space model.

### The regression coefficients are

```python
sqft_model.get('coefficients')
```

# Build a more complex model by including more features

```python
my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
```

```python
data[my_features].show()
```

```python
my_features_model = graphlab.linear_regression.create(training,target='price',features=my_features,validation_set=None)
```

## Compare two models

```python
print sqft_model.evaluate(testing)
print my_features_model.evaluate(testing)
```

## This shows that the RMSE goes down from \$263782 to \$189417 with more features.

```python
plt.plot(testing['sqft_living'],testing['price'],'*',
        testing['sqft_living'],sqft_model.predict(testing),'-',
        testing['sqft_living'], my_features_model.predict(testing),'-')
```

## Buil a regression model with more features

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house
'grade', # measure of quality of construction
'waterfront', # waterfront property
'view', # type of view
'sqft_above', # square feet above ground
'sqft_basement', # square feet in basement
'yr_built', # the year built
'yr_renovated', # the year renovated
'lat', 'long', # the lat-long of the parcel
'sqft_living15', # average sq.ft. of 15 nearest neighbors
'sqft_lot15', # average lot size of 15 nearest neighbors 
]
```

```python
my_advanced_features_model = graphlab.linear_regression.create(training,target='price',features=advanced_features,validation_set=None)
```

```python
print my_features_model.evaluate(testing)
print my_advanced_features_model.evaluate(testing)
```
